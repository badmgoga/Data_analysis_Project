```python
def GetNews(url0):
  page0 = requests.get(url0)
  soup0 = BeautifulSoup(page0.text, 'lxml')
  author = soup0.find_all('meta', {'name': 'author'})[0].attrs['content']
  date = soup0.find_all('meta', {'itemprop': 'datePublished'})[0].attrs['content']
  title = soup0.find_all('meta', {'property': 'og:title'})[0].attrs['content']
  return url0, date, author, title
```

```python
from time import sleep
news = []

for link in urls:
  res = GetNews(link)
  news.append(res)
news[0]
```

```python
import pandas as pd
df = pd.DataFrame(news)
df
```

```python
df.columns = ["link", "date", "author", "title"]
df
```

```python
df.to_excel("nplus_news.xlsx")
```

```python
pdf_attributes = soup.find_all('a', {"data-hse-file": "PDF"})
pdf_links = [attr.attrs["href"] for attr in pdf_attributes]
pdf_links
```

```python
pdf_attributes = soup.find_all('a', {"data-hse-file": "PDF"})
pdf_links = [f"https://www.hse.ru{attr.attrs['href']}" for attr in pdf_attributes]
pdf_links
```

# Задача 3
Соберём информацию о преподавателях образовательной программы "Программная инженерия" в НИУ ВШЭ, ведущих занятия в 2023/2024 учебном году (https://www.hse.ru/ba/se/tutors). Сформируем файл MS Excel содержащий данные о ФИО, должности и языках, которыми владеет преподаватель.

```python
a = requests.get("https://www.hse.ru/ba/se/tutors")

soup = BeautifulSoup(a.text, "html.parser")

instructors = [] # сюда будем 'складывать' всех преподавателей
# print(soup.prettify) надо ведь посмотреть, как html выглядит
for item in soup.find_all("div", class_="b-greetings__item b-greetings__tutor"): # пробежимся по каждому найденному классу, где хранится информация о преподавателе
    person_data = item.find("div", class_="b-greetings__person_data small")
    name_element = person_data.find("a", class_="b link_dark2 link")
    position_element = person_data.find("div", style="margin-top: 1em;")
    if name_element and position_element:
        name = name_element.text.strip()
        position = position_element.text.strip()
        language = []
        tutor_link = person_data.find('a', class_='link_dark2')
        if tutor_link:
          link = f'https://www.hse.ru{tutor_link["href"]}{name}'
          b = requests.get(link)
          soup1 = BeautifulSoup(b.text, "html.parser")
          language_info = soup1.find(class_='main-list-language-knowledge-level')
          if language_info:
            language_dd_elements = language_info.find_all('dd')
            language = [i.text.strip() for i in language_dd_elements]
        if len(language) == 0: # если человек не владеет никаким иностранным - то он владеет только русским
          language = ["русский"]
        instructors.append({
            "Name": name,
            "Position": position,
            "Languages" : language
        })


import pandas as pd
df = pd.DataFrame(instructors)
# df
df.to_excel("instructors_hse.xlsx") # программа отрабатывает за +- 4 минуты
```
